### LabelEncoder vs OneHotEncoder ###
- Generally we **don't want a relation between input columns**
- Which means input columns are **Not Correlated to each other**
- Which means input columns are **Independent(X,Y)** to each other
- Which means input columns are **Perpendicular** to each other
- Which means input columns are **Orthogonal** to each other
- Which means input columns are **90 Degrees Phase Shift** to each other
- OneHotEncoder will achive this
- But OutHotEncoder creates new columns
    - Columns are increase means dimenssions are increase
    - More Dimenssion data need More Processing Time
    - More Dimenssion data need More Infra
    - This is called **Curse of Dimensionality**
