## To create a FOLDER in CWD and save the FILE in that FOLDER ##

```python
import os
cwd = os.getcwd()
folder_name = '01_frequency_tables'
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)

for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    df = pd.DataFrame(zip(keys,values),
                     columns=['Class','Frequency'])
    file_name = f'{i}_df.csv'
    path = os.path.join(cwd,folder_name,file_name)
    df.to_csv(path,index=False)
```

## To create a FOLDER in CWD and save the BAR CHART in that FOLDER ##

```python
import os
cwd = os.getcwd()
folder_name = '02_Bar charts'
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)

for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    plt.figure(figsize=(10,3))
    plt.bar(keys,values)
    plt.xlabel = 'Class'
    plt.ylabel = 'Frequency'
    file_name = f'{i}_df.png'
    path = os.path.join(cwd,folder_name,file_name)
    plt.savefig(path)
    plt.show()
```

## To create a FOLDER in CWD and save the PIE CHART in that FOLDER ##

```python
import os
cwd = os.getcwd()
folder_name = '03_Pie charts'
try:
    os.makedirs(folder_name)
except Exception as e:
    print(e)

for i in cat[1:]:
    keys = visa_df[i].value_counts().keys()
    values = visa_df[i].value_counts().values
    plt.pie(values,
           labels=keys)
    file_name = f'{i}_df.png'
    path = os.path.join(cwd,folder_name,file_name)
    plt.savefig(path)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

plt.figure(figsize=(12, 12))

for i in range(1, len(cat)):  # You can change to range(len(cat)) if you want all
    col = cat[i]
    plt.subplot(3, 3, i)

    value_counts = visa_df[col].value_counts()
    keys = value_counts.index
    values = value_counts.values

    # Generate colors using a colormap
    cmap = cm.get_cmap('Set3', len(keys))  # Try 'tab10', 'Set1', 'Set3', etc.
    colors = [cmap(j) for j in range(len(keys))]

    plt.bar(keys, values, color=colors)
    plt.title(col)
    plt.xticks(rotation=45)
    plt.tight_layout()
```
